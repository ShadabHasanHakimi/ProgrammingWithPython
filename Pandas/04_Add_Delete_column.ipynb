```python
import pandas as pd
import numpy as np
my_df = pd.read_csv("dog_data_short.csv")
my_df
```

```python
# Add column from list
gender = ["Male", "Female", "Male", "Male", "Female"]
my_df["Gender"] = gender
my_df
```

```python
# Add default values
my_df["Alive/Dead"] = [True]*len(my_df)
my_df
```

```python
# Add null value
my_df["Show Dog"] = [np.nan]*len(my_df)
my_df
```

```python
# add column with .insert() - Allows position
my_df.insert(1, "Adopted", [True]*len(my_df), True)
my_df
```

```python
# Add column with .assign() - creates new df
my_df2 = my_df.assign(Horse=[False]*len(my_df))
my_df2
```

```python
# Remove column
# my_df.drop("Gender", axis=1) # axis = 0 for row (By Default)
# By default the column is not deleted permanently, for deleting permanently use the following
my_df.drop("Gender", axis=1, inplace=True)
my_df
```
