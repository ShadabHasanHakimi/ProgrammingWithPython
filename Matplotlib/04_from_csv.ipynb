```python
%matplotlib ipympl
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

# setting style
plt.style.use("fivethirtyeight")
```

```python
# Counter is used to count occurence
from collections import Counter
```

```python
# importing data from csv
data = pd.read_csv("data.csv")

language_counter = Counter()
```

```python
for i in range(0, 87568):
    language_counter.update(data.loc[i, "LanguagesWorkedWith"].split(";"))

language_counter
```

```python
# for printing 15 most common languages
language_counter.most_common(15)
```

```python
# splitting languages and count in different list
languages = []
popularity = []
for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
```

```python
languages
```

```python
popularity
```

```python
plt.bar(languages, popularity)
```

```python
# name of the languages are not visible in vertical bar chart, so we will use horizontal bar chart instead
plt.title("Most popular Languages")
plt.xlabel("Number of people use")
plt.barh(languages, popularity)
plt.tight_layout()
plt.show()
```
