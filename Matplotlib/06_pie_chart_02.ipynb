```python
%matplotlib ipympl
from matplotlib import pyplot as plt
```

```python
# we should not use very much data in pie charts as it will create confusion
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
```

```python
# plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'})
explode = [0, 0, 0, 0.2, 0]
# explode will highlight the fourth element (python) by moving it outwards
# start angle will change the angle of the pie chart
# for specifying the percent we can use autopct and pass format string for percent
plt.pie(slices, labels=labels, explode = explode, autopct='%1.1f%%', wedgeprops={'edgecolor': 'black'}, shadow=True, startangle=90)
plt.title("Pie Chart")
plt.tight_layout()
plt.show()
```
